```python
from pycaret.datasets import get_data
from pycaret.classification import *
myDataSet = pd.read_csv("train.csv")
myDataSet
setup(data=myDataSet, target='NObeyesdad', train_size = 0.7, verbose=False)
cm = compare_models()
```
